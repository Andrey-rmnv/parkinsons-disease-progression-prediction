## 3.2 Protein data analysis

```python
# Calculate the coefficient of variation (CV) for NPX per patient_ids and UniProt
proteins_agg = proteins[['patient_id', 'UniProt', 'NPX']]
proteins_agg = proteins_agg.groupby(['patient_id', 'UniProt'])['NPX'].aggregate(['mean', 'std'])
proteins_agg['CV_NPX[%]'] = proteins_agg['std'] / proteins_agg['mean'] * 100

# Mean CV value for UniProt
NPX_cv_mean = proteins_agg.groupby('UniProt')['CV_NPX[%]'].mean().reset_index()
NPX_cv_mean = NPX_cv_mean.sort_values(by='CV_NPX[%]', ascending=False).reset_index(drop=True)
protein_cv_top5 = NPX_cv_mean[:5]['UniProt']

proteins_agg_top5 = proteins_agg.query('UniProt in @protein_cv_top5').reset_index()
proteins_agg_top5

sns.violinplot(x="UniProt", y="CV_NPX[%]", data=proteins_agg_top5)

plt.title('Coefficient of variation of NPX for Top 5 proteins')
# Show the plot
plt.show()
```

```python
proteins_medication = pd.merge(
    proteins,
    clinical_data_copy[['visit_id', 'upd23b_clinical_state_on_medication']],
    on='visit_id',
)
proteins_medication_top5 = proteins_medication.query('UniProt in @protein_cv_top5')
proteins_medication_top5


# Create figure and axes
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(12, 12))
column_titles = ['Medication: "On"', 'Medication: "Off" or "None"']

for i, medication in enumerate([['On'], ['Off', 'Null']]):
    proteins_medication_top5_midication = proteins_medication_top5.query('upd23b_clinical_state_on_medication in @medication')
    for j, protein in enumerate(protein_cv_top5):
        # Create a box plot
        protein_medication_NPX = proteins_medication_top5_midication.query('UniProt == @protein')
        sns.boxplot(data=protein_medication_NPX,
                    y="NPX",
                    x="visit_month", 
                    ax=axes[j,i])
        sns.stripplot(data=protein_medication_NPX, 
                      y="NPX",
                      x="visit_month", 
                      jitter=True, color='black', size=2,
                      ax=axes[j,i])

        # Add labels and title
        axes[j,i].set_xlabel("visit_month")
        axes[j,i].set_ylabel(protein + ' / NPX')
    axes[0,i].set_title(column_titles[i])

# Show plot
plt.tight_layout()
plt.show()
```

## 3.3 Peptides data analysis

```python
# Calculate the coefficient of variation (CV) for NPX per patient_ids and UniProt
peptides_agg = peptides[['patient_id', 'Peptide', 'PeptideAbundance']]
peptides_agg = peptides_agg.groupby(['patient_id', 'Peptide'])['PeptideAbundance'].aggregate(['mean', 'std'])
peptides_agg['CV_PeptideAbundance[%]'] = peptides_agg['std'] / peptides_agg['mean'] * 100

# Mean CV value for UniProt
PepAbun_cv_mean = peptides_agg.groupby('Peptide')['CV_PeptideAbundance[%]'].mean().reset_index()
PepAbun_cv_mean = PepAbun_cv_mean.sort_values(by='CV_PeptideAbundance[%]', ascending=False).reset_index(drop=True)
PepAbun_cv_top5 = PepAbun_cv_mean[:5]['Peptide']

peptides_agg_top5 = peptides_agg.query('Peptide in @PepAbun_cv_top5').reset_index()

sns.violinplot(x="Peptide", y="CV_PeptideAbundance[%]", data=peptides_agg_top5)
plt.xticks(rotation=45)
plt.title('Coefficient of variation of PeptideAbundance for Top 5 Peptides')
# Show the plot
plt.show()
```

```python
peptides_medication = pd.merge(
    peptides,
    clinical_data_copy[['visit_id', 'upd23b_clinical_state_on_medication']],
    on='visit_id',
)
peptides_medication_top5 = peptides_medication.query('Peptide in @PepAbun_cv_top5')


# Create figure and axes
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(12, 12))
column_titles = ['Medication: "On"', 'Medication: "Off" or "None"']

for i, medication in enumerate([['On'], ['Off', 'Null']]):
    peptides_medication_top5_midication = peptides_medication_top5.query('upd23b_clinical_state_on_medication in @medication')
    for j, peptide in enumerate(PepAbun_cv_top5):
        # Create a box plot
        peptides_medication_PepAbun = peptides_medication_top5_midication.query('Peptide == @peptide')
        sns.boxplot(data=peptides_medication_PepAbun,
                    y="PeptideAbundance",
                    x="visit_month", 
                    ax=axes[j,i])
        sns.stripplot(data=peptides_medication_PepAbun, 
                      y="PeptideAbundance",
                      x="visit_month", 
                      jitter=True, color='black', size=2,
                      ax=axes[j,i])

        # Add labels and title
        axes[j,i].set_xlabel("visit_month")
        axes[j,i].set_ylabel(protein + ' / NPX')
    axes[0,i].set_title(column_titles[i])

# Show plot
plt.tight_layout()
plt.show()
```

# 4. Training

## 4.1 Data preparation

Since we need to predict updrs scores on 0, 6, 12, and 24 months ahead for each visit_id, the first step is to add these values in the clinical_data.

```python
clinical_data_patient = clinical_data.query('patient_id in [55, 942]').copy()
clinical_data_patient

for plus_month in [6, 12, 24]:
    clinical_data_patient_shift = clinical_data_patient[['patient_id', 'visit_month', 'updrs_1','updrs_2','updrs_3','updrs_4']].copy()
    clinical_data_patient_shift['visit_month'] -= plus_month
    clinical_data_patient_shift.rename(columns={f'updrs_{i}': f'updrs_{i}_plus_{plus_month}' for i in range(1,5)}, inplace=True)
    clinical_data_patient = pd.merge(clinical_data_patient, clinical_data_patient_shift, on=['patient_id', 'visit_month'], how='left')

clinical_data_patient
```

```python
clinical_data_extended = clinical_data.copy()
print(clinical_data_extended.shape)

for plus_month in [6, 12, 24]:
    clinical_data_extended_shift = clinical_data_extended[['patient_id', 'visit_month', 'updrs_1','updrs_2','updrs_3','updrs_4']].copy()
    clinical_data_extended_shift['visit_month'] -= plus_month
    clinical_data_extended_shift.rename(columns={f'updrs_{i}': f'updrs_{i}_plus_{plus_month}' for i in range(1,5)}, inplace=True)
    clinical_data_extended = pd.merge(clinical_data_extended, clinical_data_extended_shift, on=['patient_id', 'visit_month'], how='left')

clinical_data_extended
```

```python
proteins_pivoted = proteins.pivot(index=['visit_id', 'visit_month', 'patient_id'], columns='UniProt', values='NPX').reset_index()
proteins_pivoted
```

```python
peptides_reduced = peptides.copy()
peptides_reduced['UniProt_Peptide'] = peptides_reduced['UniProt'] + '__' + peptides_reduced['Peptide']
peptides_reduced = peptides_reduced.drop(columns=['UniProt', 'Peptide'])
peptides_reduced

peptides_pivoted = peptides_reduced.pivot(index=['visit_id', 'visit_month', 'patient_id'], columns='UniProt_Peptide', values='PeptideAbundance').reset_index()
peptides_pivoted
```

## 4.2 Random Forest Model

Here we make a Random Forest Model to obtain feature importance scores

```python
from sklearn.impute import SimpleImputer

imp_median = SimpleImputer(strategy='median')

df_imp_median = pd.DataFrame(imp_median.fit_transform(proteins_pivoted), columns=proteins_pivoted.columns)
proteins_mednans = proteins_pivoted.combine_first(df_imp_median)

df_imp_median = pd.DataFrame(imp_median.fit_transform(peptides_pivoted), columns=peptides_pivoted.columns)
peptides_mednans = peptides_pivoted.combine_first(df_imp_median)
```
